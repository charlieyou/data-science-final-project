```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import matplotlib
matplotlib.style.use('ggplot')
```

```python
df = pd.read_csv('debate_results.csv')
```

```python
c_total = df['total'][0:3].sum()
c_pos = df['positive'][0:3].sum()
t_total = df['total'][3:6].sum()
t_pos = df['positive'][3:6].sum()
```

```python
ind = np.arange(2)
bot = (c_pos, t_pos)
top = (c_total - c_pos, t_total - t_pos)

p1 = plt.barh(ind, bot, color='g')
p2 = plt.barh(ind, top, left=bot, color='r')

plt.xlabel('Number of Comments')
plt.title('Sentiment During Debates')
plt.yticks(ind + .4, ('Clinton', 'Trump'))

plt.show()
```
